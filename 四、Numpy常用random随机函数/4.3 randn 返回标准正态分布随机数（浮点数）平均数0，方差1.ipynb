## 4.3 randn 返回标准正态分布随机数（浮点数）平均数0，方差1

randn生成一个从标准正态分布中得到的随机标量，标准正态分布即N(0，1)。
randn（n)和randn(a1,a2,...)的用法和rand相同


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams["font.sans-serif"] = ["KaiTi"]
mpl.rcParams["axes.unicode_minus"] = False


mu = 165.0
sigma = 10
count = 5000
x = mu + sigma*np.random.randn(500)

bins = 50

fig, ax = plt.subplots(1, 1)

n, bins, patches = ax.hist(x, bins, density=True, histtype="bar", facecolor="#99FF33", edgecolor="#00FF99", alpha=0.75)

y = ((1/(np.sqrt(2*np.pi)*sigma))*np.exp(-0.5*np.power((bins-mu)/sigma, 2)))

ax.plot(bins, y, color="#7744FF", ls="--", lw=2)
print(bins)
print(y)
plt.show()
```
