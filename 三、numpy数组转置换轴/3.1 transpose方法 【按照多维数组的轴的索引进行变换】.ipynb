## 3.1 transpose方法 【按照多维数组的轴的索引进行变换】

transpose 的原理其实是根据维度（shape）索引决定的

将对应的索引以及维度进行调整

```python
import numpy as np
arry=np.arange(24).reshape((4,6))
print(arry)
print("-"*30)
print(arry.transpose())
print("-"*30)
print(arry.T)
```

**三维的情况**

对其按轴进行变换


```python
x = np.arange(18).reshape((2,3,3))
print(x)
print('-'*20)
x = np.transpose(x,(1,0,2))  
print(x)
print(x.shape)
```
